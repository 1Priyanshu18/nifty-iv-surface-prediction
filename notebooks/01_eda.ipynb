```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv(r"C:\DS-Projects\iv-surface-prediction\data\raw\dataset.csv")
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['datetime']=pd.to_datetime(df['datetime'],dayfirst=True)
df.head()
```

```python
iv_cols = [c for c in df.columns if c not in ['datetime','underlying_price']]
ce_cols = [c for c in iv_cols if c.endswith('CE')]
pe_cols = [c for c in iv_cols if c.endswith('PE')] # separating call and put cols because both have different behaviour for IV.

ce_strikes = [int(c[12:-2]) for c in ce_cols]
pe_strikes = [int(c[12:-2]) for c in pe_cols]
print(ce_strikes)
print(pe_strikes)
```

### Interpretation

```python
print(f"call strike range: {min(ce_strikes)} - {max(ce_strikes)}")
print(f"put strike range: {min(pe_strikes)} - {max(pe_strikes)}")
print(f"underlying price range: {df['underlying_price'].min():.2f} - {df['underlying_price'].max():.2f}")
print(f"underlying price mean: {df['underlying_price'].mean():.2f}")
print(f"Trading days: {df['datetime'].dt.date.nunique()}")
```

```python
# underlying price movement

plt.plot(df['datetime'],df['underlying_price'])
plt.title('Underlying price vs time')
plt.xlabel('Time')
plt.ylabel('Underlying price')
plt.grid()

plt.savefig('../outputs/1_eda_figures/underlying_price_analysis.png',dpi=300,bbox_inches='tight')

plt.show()
```

### Missingness analysis

```python
total_missing = df[iv_cols].isna().sum().sum()
print(f"Missing no. of iv cells: {total_missing}")
print(f"Overall missing pct: {(total_missing/df[iv_cols].size)*100:.2f}")

ce_missing = df[ce_cols].isna().sum().sum()
pe_missing = df[pe_cols].isna().sum().sum()

print(f"pe missing pct: {(pe_missing/df[pe_cols].size)*100:.2f}")
print(f"ce missing pct: {(ce_missing/df[ce_cols].size)*100:.2f}")
```

```python
# column-wise missing analysis

miss_df = pd.DataFrame({
    'missing_count': df[iv_cols].isna().sum(),
    'missing_pct': (df[iv_cols].isna().mean())*100,
    'type': ['CE' if c.endswith('CE') else 'PE' for c in iv_cols]
})

miss_df
```

```python
plt.bar(range(len(iv_cols)),(df[iv_cols].isna().mean())*100)
plt.xlabel('Optins columns')
plt.ylabel('Missing pct')
plt.title('Missing IV pct vs strikes')

plt.savefig('../outputs/1_eda_figures/missingness_analysis_across_strikes.png')

plt.show()
```

```python
# day-wise temporal missingness analysis

daily_miss_df = df.groupby(df['datetime'].dt.date)[iv_cols].apply(lambda x: x.isna().mean().mean()*100)

daily_miss_df
```

```python
daily_miss_df.plot(kind='bar')
plt.title('Daily miss pct analysis')
plt.ylabel('miss pct')

plt.savefig('../outputs/1_eda_figures/daily_miss_analysis.png')

plt.show()
```

```python
# hour-wise temporal missingness analysis

hour_miss_df = df.groupby(df['datetime'].dt.hour)[iv_cols].apply(lambda x: x.isna().mean().mean()*100)

hour_miss_df
```

```python
hour_miss_df.plot(kind='bar')
plt.title('Hourly miss pct analysis')
plt.ylabel('miss pct')

plt.savefig('../outputs/1_eda_figures/hourly_miss_analysis.png')

plt.show()
```

### IV Surface Visualization

```python
fig, axes = plt.subplots(1,2)

sample_rows = df[(df['datetime'].dt.hour == 11) & (df['datetime'].dt.minute == 30)]

# CE Smile

for _,row in sample_rows.iterrows():

    spot_price = row['underlying_price']

    moneyness, ivs = [],[]

    for i, col in enumerate(ce_cols):
        if(pd.notna(row[col])):
            strike = ce_strikes[i]
            moneyness.append(strike/spot_price)
            ivs.append(row[col])

    axes[0].plot(moneyness,ivs,marker='o',label=str(row['datetime'].date()))

axes[0].set_title('CE IV Smile')
axes[0].set_xlabel('Moneyness')
axes[0].set_ylabel('IV')
axes[0].legend()
axes[0].grid(True)


# PE Smile

for _,row in sample_rows.iterrows():

    spot_price = row['underlying_price']

    moneyness, ivs = [],[]

    for i,col in enumerate(pe_cols):
        if(pd.notna(row[col])):
            strike = pe_strikes[i]
            moneyness.append(strike/spot_price)
            ivs.append(row[col])

    axes[1].plot(moneyness,ivs,marker='o',label=str(row['datetime'].date()))

axes[1].set_title('PE IV Smile')
axes[1].set_xlabel('Moneyness')
axes[1].set_ylabel('IV')
axes[1].legend()
axes[1].grid(True)

plt.show()

plt.savefig('../outputs/1_eda_figures/IV_smile.png')
    
```

```python
for col in iv_cols:
    if df[col].max() > 1:
        idx = df[col].idxmax()

        print(
            col,
            df.loc[idx, 'datetime'],
            df[col].max()
        )
```

##### The market prices higher uncertainty in deep OTM options and Tail events which creates the classic volatility smile.
##### Near expiry: Option premium becomes very small, Tiny price changes produce huge IV changes, Deep OTM options become numerically unstable. Hence IV explodes in the wings. This is a well-known phenomenon: Volatility Explosion in the Wings

### IV distribution

```python
ce_iv = df[ce_cols].values.flatten()
ce_iv = ce_iv[~np.isnan(ce_iv)]

pe_iv = df[pe_cols].values.flatten()
pe_iv = pe_iv[~np.isnan(pe_iv)]

plt.hist(ce_iv,label='CE IVs',bins=50,alpha=0.6)
plt.hist(pe_iv,label='PE IVs',bins=50,alpha=0.6)

plt.title('Distribution of PE and CE IVs')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig('../outputs/1_eda_figures/IV_distribution.png')
```

### ATM IV temporal

```python
atm_ce_iv, atm_pe_iv = [],[]

for _,row in df.iterrows():
    spot_price = row['underlying_price']

    nearest_ce = min(ce_strikes,key = lambda strike: abs(strike-spot_price)) 
    nearest_pe = min(pe_strikes,key = lambda strike: abs(strike-spot_price))

    atm_ce_iv.append(row[f"NIFTY27JAN26{nearest_ce}CE"])
    atm_pe_iv.append(row[f"NIFTY27JAN26{nearest_pe}PE"])

df['atm_ce_iv'] = atm_ce_iv
df['atm_pe_iv'] = atm_pe_iv

plt.plot(df['datetime'],df['atm_ce_iv'],label='CE ATM temporal evolution')
plt.plot(df['datetime'],df['atm_pe_iv'],label='PE ATM temporal evolution')
plt.xlabel('Time')
plt.ylabel('IV')
plt.legend()

plt.show()

plt.savefig('../outputs/1_eda_figures/ATM_evolution.png')
```
